```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Deltas = [0.5, 1.0, 1.5]
Gaps = [-0.5, 0.0, 0.5]
Gammas = [0.003, 0.005, 0.008]
for Delta in Deltas:
    for gap in Gaps:
        data = np.load('results/gapgamma_Delta{}_gap{}.npz'.format(Delta, gap))
        Ts = data['Ts']
        mus = data['mus']
        S_boltz = data['Seebeck_Boltz']
        S_kubo = data['Seebeck_Kubo']
        Ngamma = S_kubo.shape[1]

        for g in range(Ngamma):
            plt.plot(Ts, S_kubo[:, g])
        #plt.show()
```

```python
Delta = Deltas[0]
gap = Gaps[0]
data = np.load('gapgamma_Delta{}_gap{}.npz'.format(Delta, gap))
Ts = data['Ts']
K0_kubo = data['K0_kubo']
K1_kubo = data['K1_kubo']
K0_boltz = data['K0_boltz']
K1_boltz = data['K1_boltz']
Ts_extend = np.repeat(Ts, 3).reshape(-1, 3)
S_kubo = - K1_kubo / K0_kubo / Ts_extend
S_boltz = - K1_boltz / K0_boltz / Ts_extend
plt.plot(Ts, S_kubo[:,0])
plt.plot(Ts, S_kubo[:,1])
plt.plot(Ts, S_kubo[:,2])
```
